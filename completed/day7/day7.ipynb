```python
def read_file(path):
    file = open(path,"r")
    lines = file.read()
    file.close()
    return lines

raw_input = read_file('input.txt')


# raw_input = '''2345A 1
# Q2KJJ 13
# Q2Q2Q 19
# T3T3J 17
# T3Q33 11
# 2345J 3
# J345A 2
# 32T3K 5
# T55J5 29
# KK677 7
# KTJJT 34
# QQQJA 31
# JJJJJ 37
# JAAAA 43
# AAAAJ 59
# AAAAA 61
# 2AAAA 23
# 2JJJJ 53
# JJJJ2 41'''
```

```python
# advent of code day 7


# raw_input = '''32T3K 765
# T55J5 684
# KK677 28
# KTJJT 220
# QQQJA 483'''


mod_input = [i.split() for i in raw_input.strip().split('\n')]

#initialize a shell for a check dict
lookup_dict = {}
check_dict = {}
for n,i in enumerate(['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']):
    check_dict.setdefault(i,0)
    lookup_dict.setdefault(i,"{:02d}".format(13-n))

count_dict = {}
for i in range(5,0,-1):
    count_dict.setdefault(i,0)

checker = check_dict.copy()
counter = count_dict.copy()


final_dict = {}
for n,i in enumerate(mod_input):
    checker = check_dict.copy()
    counter = count_dict.copy()
    for j in i[0]:
        checker[j] += 1
    checker_id = int(''.join([str(i) for i in list(checker.values())]))
    val = ''
    for l in i[0]:
        val += str(lookup_dict[l])
        

    for k in list(checker.values()):
        if k != 0:
            counter[k] +=1
    counter_id = int(''.join([str(i) for i in list(counter.values())]))

    final_dict[n] = {'hand':i[0],'bet':i[1],'checker_id':checker_id,'counter_id':counter_id,'value':int(val)}

final_dict = dict(sorted(final_dict.items(),key=lambda x: (x[1]['counter_id'],x[1]['value']),reverse = False))

total = 0
for n,i in enumerate(final_dict.items()):
    step = n+1
    bet = int(i[1]['bet'])
    total += step*bet

print(total)
```
