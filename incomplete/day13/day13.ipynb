```python
def check_pattern(fields:list)->list:
    max_pattern = 0
    input_type = 0
    max_coords = []


    for input in [fields,hf.transpose(fields)]:
        

        i=0
        while i < len(input)-1:
            if input[i] == input[i+1]:

                pattern = True
                pattern_search_iter = 1
                while pattern:
                    if (i-pattern_search_iter <0) or (i+1+pattern_search_iter > len(input)-1):
                        max_coords = [i,i+1]
                        return [i,i+1],input_type
                    elif input[i-pattern_search_iter] == input[i+1+pattern_search_iter]:
                        pattern_search_iter+=1
                    else:
                        pattern = False     
            i+=1
        input_type += 1

    return max_coords, input_type







total = 0

for i in final_input:
    coords, input_type = check_pattern(i)
    if input_type ==1: #vertical
        total += coords[0]+1
    else: #horizontal
        total += (coords[0]+1)*100

print(total)
#46311 too high
#32532 too high
```

```python
#part 2

def single_string_diff(string1:str,string2:str):
    count = 0
    occurance = []
    for i in range(len(string1)):
        if string1[i] != string2[i]:
            count += 1
            occurance.append(i)
    
    if count == 1:
        return occurance[0]
    else:
        return None
    
single_string_diff('..##..##.','#.##..##.')


```
