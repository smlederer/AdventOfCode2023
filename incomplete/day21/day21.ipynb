```python
#part 2:

raw_input = '''...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
...........'''

# raw_input = hf.read_file('input.txt')

mod_input = hf.modify_raw(raw_input)
M = hf.matrix(mod_input)
offset = M.dimension
M_copy = hf.matrix(mod_input)
M_copy.set_cell(*M_copy.find('S')[0],'.')
M_copy.nice_print()

starting_position = M.find('S')[0]

queue = [[starting_position]]

iter = 0

def move2(matrix:hf.matrix,x:int,y:int)->list:
    data = matrix.get_neighbor_content(x,y,clamped=False)
    for i in data:
        mark(matrix,*i['coord'])
    return None


def mark(matrix:hf.matrix,x:int,y:int)->list:
    content = matrix.get_cell(x,y)
    if content == '.':
        matrix.set_cell(x,y,'O')
    else:
        pass
    return None


def check_bounds(matrix:hf.matrix,x,y):
    x = int(x)
    y = int(y)
    if (x<0):
        return 'l'
    elif (x>matrix.dimension[0]):
        return 'r'
    elif (y<0):
        return 'u'
    elif (y>matrix.dimension[1]):
        return 'd'
    else:
        return None


def adjust_queue(queue:list,dir:str,offset:list)->list:
    new_queue = []
    for i in queue:
        leaf = []
        for j in i:
            if dir == 'l':
                leaf.append(hf.vector_op(j,(offset[0],0),'+'))
            elif dir == 'u':
                leaf.append(hf.vector_op(j,(0,offset[1]),'+'))

        new_queue.append(leaf)
    return new_queue


break_flag = False
while len(queue)>0:

    for i in queue[0]:
        if not break_flag:
            #find the neighbors of the move
            data = M.get_neighbor_content(*i,clamped=False)
            for point in data:
                bound_flag = check_bounds(M,*point['coord'])
                if bound_flag:
                    M.append(M_copy.matrix,bound_flag)
                    #translate all points in queue if u or l:
                    if bound_flag in ('l','u'):
                        print(queue)
                        queue = adjust_queue(queue,bound_flag,hf.vector_op(vector1=offset,vector2=[-1,-1],operation='+'))
                        print(queue)
                break_flag = True
    if not break_flag:       
        for i in queue[0]:
            move(M,*i)


        next_coords = M.find('O')
        queue.pop(0)
        queue.append(next_coords)

        iter += 1

        if iter ==18:
            queue = []
    break_flag = False

M.nice_print()
```
