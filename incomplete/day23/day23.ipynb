```python
M_new = hf.matrix(None,*M.dimension)

def pathfind(matrix:hf.matrix,start,end)->list:
    err={}
    queue = []
    init_dict =  {'current':start,'visited':[],'path':[]}
    queue.append(init_dict)
    while len(queue)>0:
        curr = queue[0]
        visited = curr['visited']
        current_cell = curr['current']
        path = curr['path']
        visited.append(current_cell)
        

        next_possible = [x['coord'] for x in matrix.get_neighbor_content(*current_cell) if x['content'] != '#' and x['coord'] not in set(visited)]
        curr_symbol = M.get_cell(*current_cell)
        if curr_symbol in ('>','v','<','^'):
            err['_'.join([str(x) for x in current_cell])] = {'path':path,'symbol':curr_symbol,'current_cell':current_cell}
            path = []
        if current_cell == end:
            err['_'.join([str(x) for x in current_cell])] = {'path':path,'symbol':'E','current_cell':current_cell}
        path.append(current_cell)
        for cell in next_possible:
            if cell not in visited:
                queue.append({'current':cell,'visited':visited,'path':path})
        # if current_cell == end:
        #     print({'current':cell,'visited':visited})


        queue.pop(0)
        # print(queue)

    return err

def test_print(coord_list:list,val)->hf.matrix:
    M_new = hf.matrix(None,*M.dimension)
    for n,i in enumerate(coord_list):
        M_new.set_cell(*i,val) 
    
    return M_new

# for n,i in enumerate(pathfind(M,(1,0),(21,22))):
#     z = test_print(i,chr(87+n))
#     print('\n'.join(z.nice_print()),'\n')

arrow_dict = {'>':[1,0],'v':[0,1],'<':[-1,0],'^':[0,-1],'E':[0,0]}

```

```python
fdict =pathfind(M,(1,0),(21,22))

for curr in fdict.keys():
    link = tuple(hf.vector_op(fdict[curr]['current_cell'],arrow_dict[fdict[curr]['symbol']],'+'))
    for other in fdict.keys():
        if curr != other:
            searching =  fdict[other]['path']
            print(link,searching)
            if link in searching:  
                fdict[curr].setdefault('connections',[])
                fdict[curr]['connections'].append(other)
```

```python
tot_path = 0
start = list(fdict.keys())[0]

queue = [{'total':tot_path,'curr':start}]


while len(queue)>0:
    node = queue[0]
    tot_path = node['total']
    tot_path += len(fdict[node['curr']]['path'])
    if fdict[node['curr']]['symbol']=='E':
        print(tot_path)
    elif 'connections' in fdict[node['curr']]:
        for i in fdict[node['curr']]['connections']:
            queue.append({'total':tot_path,'curr':i})
    queue.pop(0)

        
```
