```python
def map_read(string:str):
    tmp = string.split('\n')
    return {tmp[0]:[i.split(' ') for i in tmp[1:]]}

input_dict = {}
for i in mod_input[1:]:
    map_name = list(map_read(i).items())[0][0]
    values = list(map_read(i).items())[0][1]
    input_dict.setdefault(map_name,values)

print(input_dict)
```

```python


def mod_map(map_values:dict) ->dict:
    source = []
    mod = []
    destination = []
    for j in map_values:
        i = [int(x) for x in j]

        destination.append((i[0],i[0]+i[2]-1))
        source.append((i[1],i[1]+i[2]-1))
        mod.append(i[0]-i[1])

    return {'source':source,'destination':destination,'mod':mod}


final_dict = {}
for i in input_dict.items():
    values = mod_map(i[1])

    final_dict.setdefault(i[0],values)

print(final_dict)

def check_seed(num:int,map_dict:dict)->int:
    for i in range(len(map_dict['source'])):
        lower_bound = map_dict['source'][i][0]
        upper_bound = map_dict['source'][i][1]
        if (num>=lower_bound) and (num <= upper_bound):
            mod_num = num + map_dict['mod'][i]
            break
        else:
            mod_num = num
    return mod_num

def translate(num):
    mod_num = num
    for i in final_dict.keys():
        mod_num = check_seed(mod_num,final_dict[i])

    return mod_num

min = None
for i in mod_input[0].split(': ')[1].split(' '):
    value = int(i)
    mod_val = translate(int(i))
    if min == None:
        min = mod_val
    elif mod_val < min:
        min = mod_val
        

print(min)

```
