```python
import sys
import os

__file__ = 'helper_funcs.py'
file_path = os.path.abspath(os.path.join(os.path.dirname( __file__ ), '..'))
sys.path.append(file_path)

import helper_funcs as hf
```

```python
# raw_input = r'''.|...\....
# |.-.\.....
# .....|-...
# ........|.
# ..........
# .........\
# ..../.\\..
# .-.-/..|..
# .|....-|.\
# ..//.|....'''

#going to have to figure out how to handle the double slash read
raw_input = hf.read_file('input.txt')
mod_input = hf.modify_raw(raw_input)
#problem 16 specific:
dic = {'-':'H','|':'V','\\':'L','/':'R'}

for i in dic.items():
    mod_input = [x.replace(i[0],i[1]) for x in mod_input]


M = hf.matrix(mod_input)
[print(len(x)) for x in M.matrix]
M.matrix
```

```python
direc_dict = {'E':[1,0],'W':[-1,0],'N':[0,-1],'S':[0,1]}

queue = [{'direc':'E','coor':[0,0]}]

M_new = hf.matrix(None,*M.dimension)

def handle_symbol(symbol:str,dir:str)->str:
    effect_dic = {'V':{'N':'no_effect','S':'no_effect','E':'split','W':'split'},
            'H':{'N':'split','S':'split','E':'no_effect','W':'no_effect'},
            'R':{'N':'E','S':'W','E':'N','W':'S'},
            'L':{'N':'W','S':'E','E':'S','W':'N'}}
    
    if symbol in ('V','H'):
        if effect_dic[symbol][dir] == 'split':
            direction = [i[0] for i in effect_dic[symbol].items() if i[1]=='no_effect'] #replace the beam with split beams
        else:
            direction = list(dir)

    elif symbol in ('R','L'):
        direction = list(effect_dic[symbol][dir])

    else:
        direction = dir

    return direction




visited = []
iter = 0
while len(queue)>0:
    current = queue[0]
    current_cell = current['coor']
    
    if (not M.clamp(*current_cell)):
        pass
    else:
        if (current_cell not in visited):
            current_dir = current['direc']
            M_new.set_cell(*current_cell,'#')
            current_symbol = M.get_cell(*current_cell)
            # if current_symbol == '.':
            #     visited.append(current_cell)
            current_effect = handle_symbol(current_symbol,current_dir)

            for i in current_effect:
                queue.append({'direc':i,'coor':hf.vector_op(current_cell,direc_dict[i],'+')})
    queue.pop(0)

    iter+=1
    if iter > 10000000:
        queue=[]
    


print(M_new.matrix)




```
